3. Make a bar-plot of distance in `km` by day. The day can be obtained from `startDate` column

````python
from datetime import datetime

df["dates"] = pd.to_datetime(df['startDate'], format='%Y-%m-%d').dt.date

df
````

````python
date_sum = df.groupby("dates")["value"].sum()
date_sum
````

````python
import matplotlib.pyplot as plt

print(date_sum.index)

fig = plt.figure(figsize=(100, 20))
plt.bar(x=date_sum.index, height=date_sum.values)
plt.grid(axis = "y")

plt.xticks(rotation=90)

plt.xlabel("Dates")
plt.ylabel("Kilometers")
````

4. Implement a game of tic-tac-toe. On each move, a random field is selected from the unoccupied ones and, depending on the order, a cross or a zero is put there. When the free fields end, you need to calculate who won. The game does not require user input. The whole procedure is calculated automatically.

When implementing, use numpty to set the field as a two-dimensional array

````python
import numpy as np
import random

free_spots = []
for i in range(3):
    for j in range(3):
        free_spots.append((i, j))

playground = np.full(shape=(3, 3), fill_value='', dtype='<U1')

for step in range(9):
    mark = 'x' if step % 2 == 0 else 'o'
    i, j = random.choice(free_spots)
    free_spots.remove((i, j))
    playground[i, j] = mark

number_wins_o = 0
number_wins_x = 0
for i in range(3):
    number_wins_o += (playground[i] == 'o').all()
    number_wins_x += (playground[i] == 'x').all()
for j in range(3):
    number_wins_o += (playground[:, j] == 'o').all()
    number_wins_x += (playground[:, j] == 'x').all()

for diagonal in (playground.diagonal(), np.fliplr(playground).diagonal()):
    number_wins_o += (diagonal == 'o').all()
    number_wins_x += (diagonal == 'x').all()

print(playground)

print(f'x wins: {number_wins_x}\no wins: {number_wins_o}\n')
````
